```python
import pandas as pd
from flask import Flask, request, jsonify
import os
from google.cloud import storage

app = Flask(__name__)

def download_file(bucket_name, source_blob_name, destination_file_name):
    """Download a file from GCP Storage."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(source_blob_name)
    blob.download_to_filename(destination_file_name)

def upload_file(bucket_name, destination_blob_name, source_file_name):
    """Upload a file to GCP Storage."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
    blob.upload_from_filename(source_file_name)

def transform_data(input_path, output_path):
    # Transformations from the uploaded Jupyter notebook
    # Load the data
    data = json.load(json_file)

    # Convert the JSON data to a DataFrame
    df = pd.json_normalize(data)

    # Save the DataFrame to a CSV file
    df.to_csv('data.csv', index=False)
    data = pd.read_csv('/content/data.csv')
    data['start_date'] = pd.to_datetime(data['start_date'])
    data['start_date_local'] = pd.to_datetime(data['start_date_local'])

    # Extract additional time-related features
    data['start_day'] = data['start_date_local'].dt.day
    data['start_hour'] = data['start_date_local'].dt.hour
    data['start_weekday'] = data['start_date_local'].dt.weekday
    data.columns = data.columns.str.replace('.', '_', regex=False)




@app.route('/', methods=['POST'])
def main():
    """Main entry point for the Cloud Run function."""
    request_json = request.get_json()
    bucket_name = request_json['bucket_name']
    source_blob_name = request_json['source_blob_name']
    destination_blob_name = request_json['destination_blob_name']

    # File paths
    input_path = '/tmp/input.csv'
    output_path = '/tmp/output.csv'

    # Download the file from GCP Storage
    download_file(bucket_name, source_blob_name, input_path)

    # Perform the transformation
    transform_data(input_path, output_path)

    # Upload the transformed file back to GCP Storage
    upload_file(bucket_name, destination_blob_name, output_path)

    return jsonify({'status': 'success'}), 200

if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=int(os.environ.get('PORT', 8080)))
```
