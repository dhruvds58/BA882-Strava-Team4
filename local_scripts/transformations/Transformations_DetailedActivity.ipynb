```python
import pandas as pd
from flask import Flask, request, jsonify
import os
from google.cloud import storage

app = Flask(__name__)

def download_file(bucket_name, source_blob_name, destination_file_name):
    """Download a file from GCP Storage."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(source_blob_name)
    blob.download_to_filename(destination_file_name)

def upload_file(bucket_name, destination_blob_name, source_file_name):
    """Upload a file to GCP Storage."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
    blob.upload_from_filename(source_file_name)

def transform_data(input_path, output_path):
    # Transformations from the uploaded Jupyter notebook
    # Load the data
    data = json.load(json_file)

    # Convert the JSON data to a DataFrame
    df = pd.json_normalize(data)

    # Save the DataFrame to a CSV file
    df.to_csv('data.csv', index=False)
    data = pd.read_csv('/content/data.csv')
    columns_to_drop = [
    "utc_offset", "heartrate_opt_out", "display_hide_heartrate_option",
    "from_accepted_tag", "has_kudoed", "description", "segment_efforts",
    "splits_metric", "splits_standard", "laps", "best_efforts",
    "stats_visibility", "hide_from_home", "available_zones",
    "athlete.resource_state", "map.id", "map.polyline", "map.resource_state",
    "map.summary_polyline", "gear.id", "gear.nickname", "gear.resource_state",
    "gear.retired", "gear.converted_distance", "photos.primary.unique_id",
    "photos.primary.urls.600", "photos.primary.urls.100", "photos.primary.source",
    "photos.primary.media_type", "photos.use_primary_photo", "photos.count",
    "similar_activities.trend.speeds", "similar_activities.trend.current_activity_index",
    "similar_activities.trend.min_speed", "similar_activities.trend.mid_speed",
    "similar_activities.trend.max_speed", "similar_activities.trend.direction",
    "similar_activities.resource_state"]

    # Drop the specified columns
    data = data.drop(columns=columns_to_drop)
    data['start_date'] = pd.to_datetime(data['start_date'])
    data['start_date_local'] = pd.to_datetime(data['start_date_local'])
    boolean_columns = ['trainer', 'commute', 'manual', 'private', 'flagged',
                   'device_watts', 'has_heartrate', 'prefer_perceived_exertion',
                   'gear.primary']

    # Convert boolean columns to 0 and 1
    data[boolean_columns] = data[boolean_columns].astype(int)
    data['elevation_change'] = data['elev_high'] - data['elev_low']
    data['day_of_week'] = data['start_date_local'].dt.day_name()  # Get the day name (e.g., Monday)
    data['hour'] = data['start_date_local'].dt.hour               # Get the hour of the day
    data['month'] = data['start_date_local'].dt.month_name()      # Get the month name (e.g., October)
    data.columns = data.columns.str.replace('.', '_', regex=False)



@app.route('/', methods=['POST'])
def main():
    """Main entry point for the Cloud Run function."""
    request_json = request.get_json()
    bucket_name = request_json['bucket_name']
    source_blob_name = request_json['source_blob_name']
    destination_blob_name = request_json['destination_blob_name']

    # File paths
    input_path = '/tmp/input.csv'
    output_path = '/tmp/output.csv'

    # Download the file from GCP Storage
    download_file(bucket_name, source_blob_name, input_path)

    # Perform the transformation
    transform_data(input_path, output_path)

    # Upload the transformed file back to GCP Storage
    upload_file(bucket_name, destination_blob_name, output_path)

    return jsonify({'status': 'success'}), 200

if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=int(os.environ.get('PORT', 8080)))
```
